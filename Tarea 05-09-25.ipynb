### Datos de personas que ven el anime "Los diarios de la boticaria"

```python
import numpy as np
import pandas as pd
import statistics as sts
import seaborn as sns
```

```python
#Importando los datos sin valores extremos
datos_normales=np.loadtxt('./tarea1.csv',delimiter=',',dtype=np.int16)
datos_normales = datos_normales.flatten()
#Importando los datos con valores extremos
datos_outliers=np.loadtxt('./tarea1_2.csv',delimiter=',',dtype=np.int16)
datos_outliers = datos_outliers.flatten()
```

```python
#Función para calcular el rango intercuartiles
def calcular_iqr(data):
    #Calcular cuartiles 1 y 3
    qs = np.quantile(data,[0.25, 0.75])

    #Calcular IQR
    iqr = qs[1] - qs[0]
    return iqr
```

#### Medidas de tendencia central y variabilidad para los datos sin outliers

```python
print("Datos sin outliers")
print(datos_normales)
```

```python
print("Datos sin outliers ordenados")
print(np.sort(datos_normales))
```

```python
#Media
print(f'Media: {datos_normales.mean()}')
#Mediana
print(f'Mediana: {np.median(datos_normales)}')
#Moda
print(f'Moda: {sts.mode(datos_normales)}')
#Desviación estándar
print(f'Desviación estándar: {datos_normales.std()}')
#Rango intercuartiles
print(f'IQR: {calcular_iqr(datos_normales)}')
```

#### Medidas de tendencia central y variabilidad para los datos con outliers

```python
print("Datos con outliers")
print(datos_outliers)
```

```python
print("Datos sin outliers ordenados")
print(np.sort(datos_outliers))
```

```python
#Media
print(f'Media: {datos_outliers.mean()}')
#Mediana
print(f'Mediana: {np.median(datos_outliers)}')
#Moda
print(f'Moda: {sts.mode(datos_outliers)}')
#Desviación estándar
print(f'Desviación estándar: {datos_outliers.std()}')
#Rango intercuartiles
print(f'IQR: {calcular_iqr(datos_outliers)}')
```

#### Conclusiones

Observando la siguiente tabla comparativa entre los datos sin y con outliers:

| | Media | Desviación Estándar | | Mediana | IQR |
|---|:---:|:---:|---|:---:|:---:|
| Sin outliers | 16.46 | 1.161 | | 16.0 | 3.0 |
| Con outliers | 21.409 | 15.828 | | 17.0 | 3.0 |

Encontramos que los datos normales tienen una media de 16.46 y cuentan con una desviación estándar de 1.161, esto quiere decir que los datos no están tan alejados de la media, y podemos observar que su mediana al ser igual a 16 podemos asumir que la media y mediana tienen mucha coherencia entre ellos, ahora el rango intercuartiles tiene un valor de 3 y esto indica que los datos se encuentran cercas del valor de la mediana.

Ahora, analizando los datos con outliers podemos encontrar que la media se dispara o tiene una gran diferencia a comparación de los que no contienen datos extremos (21.409 vs 16.46), esto quiere decir que los outliers influyeron demasiado en la media, lo mismo pasa con la desviación estándar (15.828 vs. 1.161), y por lo tanto podemos encontrar que los outliers están afectando a la distancia que hay entre los datos y el valor de la media; en cuanto a la mediana podemos observar que tiene una ligera diferencia a comparación de los datos normales (17 vs. 16) y el IQR se mantiene igual (3 vs. 3) a pesar de que estos datos tienen valores extremos.

Con esta información podemos concluir que los valores extremos afectan en gran medida a la obtención de la media y desviación estándar de los datos, si queremos tener datos más fiables respecto a nuestra información es mejor utilizar la mediana y el IQR debido a que a pesar de que existen outliers, no se ven exageradamente afectados por esto, es decir, que tienen una variación muy mínima o como en este caso nula si comparamos los valores del IQR.

### Puntuación Z

Obtener el porcentaje de probabilidad en el rango de 16 a 17 años

Para $X=16$

$$Z=\frac{X-\mu}{\sigma}$$
$$Z=\frac{16-16.46}{1.161206}$$
$$Z=\frac{-0.46}{1.161206}$$
$$Z=-0.396139$$

$$\phi(-0.39) \approx 0.34827$$

Para $X=17$

$$Z=\frac{X-\mu}{\sigma}$$
$$Z=\frac{17-16.46}{1.161206}$$
$$Z=\frac{0.54}{1.161206}$$
$$Z=0.465$$

$$\phi(0.46) \approx 0.67724$$

Entonces:
$$P(16 \leq X \leq 17)=\phi (0.46)-\phi (-0.39)$$
$$P(16 \leq X \leq 17)=0.67724-0.34827$$
$$P(16 \leq X \leq 17)=0.32897 \approx 0.329 \rightarrow 32.9%$$
