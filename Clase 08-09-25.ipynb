## Gráfico de caja

El análisis exploratorio es una pieza fundamental en la Minería de Datos, los gráficos de caja (boxplots) nos permiten un análisis univariado de variables numéricas en el análisis exploratorio de datos. Nos permitirá ver lo siguiente:

- Valores centrales
- Dispersión de los datos
- Simetría
- Outliers

![Descripción de la imagen](https://mathsmadeeasy.co.uk/wp-content/uploads/2018/11/boxplots-figure-1-e1597247984531.png)

Sus elementos son:

- Los cuartiles 1 y 3 (limites inferior y superior de la caja)
- La mediana (línea central de la caja)
- Los bigotes (whiskers)
  - El inferior es simplemente *Q1-(1.5xIQR)*
  - El superior es simplemente *Q3+(1.5xIQR)*

Nota: Se puede usar 3 en lugar de 1.5 en caso de que los valores sean muy extremos

Si se necesita ver una distribución más detallada de los datos:

- Podemos ver la tendencia central (mediana) y la dispersión (IQR)
- Contamos con simetría o asimetría en la distribución
- Si hay valores extremos (outliers)
- Dentro de qué rango de valores se encuentra la mayor parte de los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def calcular_iqr(data):
    #Calcular cuartiles 1 y 3
    qs = data.quantile([0.25, 0.75])

    #Calcular IQR
    iqr = qs[0.75] - qs[0.25]
    return iqr, qs[0.25], qs[0.75]
```

```python
df = pd.read_csv('./bigotes.csv')
df
```

```python
print('A:')

IQR,Q1,Q3 = calcular_iqr(df['A'])
print(f'\tMedia: {df['A'].mean()}')
print(f'\tMediana: {df['A'].median()}')
print(f'\tDesviación Estándar: {df['A'].std()}')
print(f'\tIQR: {IQR}')
print(f'\tQ1: {Q1}')
print(f'\tQ3: {Q3}')
print(f'\tMin: {df['A'].min()}')
print(f'\tMax: {df['A'].max()}')
print(f'\tBigote Min: {Q1-(1.5*IQR)}')
print(f'\tBigote Max: {Q3+(1.5*IQR)}')
```

Mientras más grande la std es porque esta más aleatorio. Estos datos están entre -1 y 7, si se salen es un outliers. Datos asimetricos es cuando la media no esta a la mitad, si los datos estan en el maximo quiere decir que la grafica esta a la izquierda

```python
#Boxplot
ax = sns.boxplot(data=df['A'], showfliers=True) #showfilters=False quitar los outliers
ax = sns.stripplot(data=df['A'], jitter=True, alpha=0.5) #poder superponer los datos
ax.set_xlabel('Variable A')
ax.set_ylabel('Valores')
ax.axhline((Q1-(1.5*IQR)), color='black', linestyle='--')
ax.axhline((Q3+(1.5*IQR)), color='black', linestyle='--')
```

La línea superior es el bigote superior, si se sale de la línea es un outlier, todos los datos del 7 para abajo son correctos, si están arriba son outliers

### Resultados

- Tenemos una mediana de 3
- IQR de 2
- Los datos están entre 1 y 5
- La distribución de los datos es simetrica, la mediana esta a la mitad de la caja, porque los bigotes están a la misma distancia
- Tenemos un valor extremo de 10
