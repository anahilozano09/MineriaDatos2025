## Medidas de tendencia central

Es la manera de cuantificar el valor más comúnmente encontrado en nuestros datos.

- Datos numéricos:
    - La media (promedio)
    - La mediana
- Datos categoricos:
    - La moda

Notación
- $\bar x$ cuando nos referimos al promedio de la muestra
- $\mu$ cuando nos referimos al promedio de la población

Si tiene una **distribución normal** nuestros datos numéricos, se debería ocupar la media

### Media o promedio

Es simplemente la suma de todos los valores dividida entre el número total de valores de la muestra 

```python
datos =[2,5,10,8]
media = sum(datos)/len(datos)
print(f'Media = {media}')
```

La limitación de la media es que es sensible a valores extremos (outliers)

#### Ejemplo

```python
datos =[2,5,10,8, 300]
media = sum(datos)/len(datos)
print(f'Media = {media}')
```

### Mediana

Si hay variables numericas y no tienen una distribución normal o cuentan con valores extremos, resulta más adecuado usar la mediana:
> La mediana es simplemente el valor medio de un conjunto de datos cuando estos están ordenados de menor a mayor

#### Ejemplo

```python
datos = np.array([2,1,6,10,5])
print(f'Datos originales: {datos}')
print(f'Datos ordenados: {np.sort(datos)}')
print(f'Datos originales: {np.median(datos)}')
```

```python
datos = np.array([2,1,6,10,5,9])
print(f'Datos originales: {datos}')
print(f'Datos ordenados: {np.sort(datos)}')
print(f'Datos originales: {np.median(datos)}')
```

```python
datos = np.array([-20,2,1,6,10,5,9,100,120])
print(f'Datos originales: {datos}')
print(f'Datos ordenados: {np.sort(datos)}')
print(f'Datos originales: {np.median(datos)}')
```

### Moda

Su uso es como mediada de tendencia central para los datos categóricos:
> La moda es simplemente el valor más frecuente en la variable categórica

#### Ejemplo:

```python
datos = np.array(['Bueno','Malo','Bueno','Regular','Bueno','Malo','Malo'])
print(datos)
```

```python
moda = sts.mode(datos)
```

```python
print("Moda: ", moda)
#Nota: cambiaron la forma de hacer esto en la nueva versión
#print("Frecuencia de la moda: ", moda.count)
```

#### Ejemplo

Dataset crimenes.csv
1. ¿Cuales son la media y la mediana de la tasa de asesinatos?
2. ¿Cuales son la media y la mediana del numero total de asesinatos?

```python
#Nota: el punto indica la carpteta actual
df = pd.read_csv('./crimenes.csv')
df
```

```python
#Si son iguales hay una distribución normal
print(f'Promedio tasa de asesinatos: {df['Murder.Rate'].mean()}')
print(f'Mediana de asesinatos: {df['Murder.Rate'].median()}')
```

promedio=mediana, tiene una distribución simétrica(es decir lado y lado de la media/mediana tenemos más o menos la misma distribución de datos).

Para obtener el **número total** de asesinatos, primero hay que calcular el número total de asesinatos por cada ciudad:

*Numero total de asesinatos = Poblacion * Tasa de asesinatos / 100.000*

```python
df['Num_asesinatos'] = (1/100000)*df['Murder.Rate']*df['Population']
df
```

```python
#Si son iguales hay una distribución normal
print(f'Promedio tasa de asesinatos: {df['Num_asesinatos'].mean()}')
print(f'Mediana de asesinatos: {df['Num_asesinatos'].median()}')
```

La media es diferente a la mediana (273 vs 162 asesinatos). Esto se debe a que las poblaciones se encuentran dentro de un rango amplio de valores, lo que al momento del cálculo afecta el número total de asesinatos, esto se debe a que tenemos **valores extremos (outliers)**

```python
df['Num_asesinatos'].describe().to_frame()
```

Por lo que podemos decir que en este caso del número de asesinatos, la medida más fiable es mediana (162 asesinatos)
