```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import warnings 
import statistics as sts
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./Amazon_bestsellers_items_2025.csv')
df
```

```python
df_Mexico = df[df['country'] == "MX"]
df_Mexico
```

```python
df_MX_Microsoft = df_Mexico[df_Mexico['product_title'].str.contains("Microsoft")]
df_MX_Microsoft
```

Podemos observar que hay 27 productos de la familia Microsoft en el top 100 de los productos mejor vendidos en México

```python
df_MX_Microsoft_Rating = df_MX_Microsoft[df_MX_Microsoft['product_star_rating']>=4.7]
df_MX_Microsoft_Rating
```

Al hacer un filtrado de los datos con ayuda del código anterior podemos notar que de los 27 productos obtenidos de la familia Microsoft, solo hay 7 que tienen una calificación mayor a 4.7

```python
df_MX_Microsoft_Rating['product_num_ratings'].describe().to_frame()
```

En la tabla superior podemos encontrar que los productos de la familia Microsoft que se encuentran en el top de productos más vendidos en México, reciben una media de comentarios o de calificaciones respecto a su calidad de 366, pero hay que tener en cuenta que esta media esta siendo afectada por valores extremos debido a que el mínimo de reseñas recibidas es de 12 mientras que el máximo es de 1141, por lo que podemos notar una gran diferencia en esto

```python
df_MX_Microsoft_Rating['product_price'] = (df_MX_Microsoft_Rating['product_price'].str.replace(r"[^0-9.]", "", regex=True).astype(float))
df_MX_Microsoft_Rating
```

```python
df_MX_Microsoft_Rating['product_price'].describe().to_frame()
```

Obtenemos la información del precio de los productos mejor calificados de Microsoft, y encontramos que a pesar de que el promedio de precios de estos productos es de $2,839.2$ pesos, sin embargo, al igual que el caso de la cantidad de reseñas de estos productos, este resultado esta siendo afectado por valores extremos, esto debido a que el precio mínimo es de $1,770$ pesos y el precio máximo es de $5,999.0$ pesos, y es notable la gran diferencia entre estos valores.
