## Tablas de frecuencia

Las tablas de frecuencia son en formato tabular, nos permite ver la distribución de una variable categórica, así como las gráficas de barras

### Ejemplo

```python
df = pd.read_csv('./prestamos.csv')
df
```

Con describe si da un numero en count que no son iguales hay perdida de información, unique se le conoce como los niveles que hay, el que más aparece (top) y la cantidad con la que aparece esa información (freq)

```python
df.describe()
```

- unique: número de niveles por cada variable
- top: la categoría que más se repite en cada variable
- freq: la frecuencia de la categoría más común en cada variable

Ahora, si tiene variables numéricas `describe()`toma en cuenta estas variables descartando categóricas

```python
df['numerica']=np.random.randint(0,50, (450961))
df
```

```python
df.describe()
```

```python
#Solo las frecuencias del dato categorico que queremos obtener
df['nivel'].value_counts()
```

```python
#AHora lo hacemos de manera ascendente
df['nivel'].value_counts(ascending=True)
```

### Ejemplo practico
- escala en proporciones (normalizado)
- en orden descendente 

Da el porcentaje que tenemos

```python
#Variables categoricas
df['estado'].value_counts(normalize=True)
```

- Casi el 71% de los créditos se encuentran activos actualmente (estado = Actual)
- Un poco más del 21% de los créditos ya ha sido pagado (estado = Totalmente pagada)
- Aproximadamente un 5% de los créditos no se pueden recuperar (estado = Cancelada (impago))
- El 2.1% restante están activos pero los deudores no los han pagado a tiempo (estado = Tarde)

Ahora con nivel

```python
tabla_nivel = df['nivel'].value_counts(normalize = True)
tabla_nivel
```

```python
print(f'Porcentaje de créditos medio-altos (B,C,D): {tabla_nivel['B':'D'].sum()}')
print(f'Porcentaje de créditos medio-bajos (E,F,G): {tabla_nivel['E':'G'].sum()}')
```

- Tenemos 7 diferentes niveles de crédito, siendo B los créditos con montos más altos y G con los montos más bajos
- Aproximadamente el 72% de los créditos otorgados son de niveles medio-alto (B,C,D)
- El 16% de los créditos otorgados son de niveles altos (A)
- Aproximadamente el 12% de los créditos restantes son de niveles medio-bajo (E,F,G)

```python
tabla_estado = df['estado'].value_counts(normalize=True)
tabla_estado
```

```python
print(f'Porcentaje de estados Actual y Totalmente pagada: {tabla_estado['Actual':'Totalmente pagada'].sum()}')
print(f'Porcentaje de estados Cancelada (impago) y Tarde: {tabla_estado['Cancelada (impago)':'Tarde'].sum()}')
```
