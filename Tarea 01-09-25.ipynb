## Datos de personas de 15 a 18 años que ven el anime La boticaria

```python
import matplotlib.pyplot as plt
import numpy as np
import statistics as sts
import scipy.stats as scst
```

```python
#lectuara de los datos
arg1=np.loadtxt('./tarea1.csv',delimiter=',',dtype=np.int16)
#print(arg1)
arg2=np.sort(np.reshape(arg1,100))
#print(arg2)
rango=arg2[99]-arg2[0]
print(rango)
#intervalos de la clase
intervalos=rango/7
b=np.arange(arg2[0],arg2[99]+intervalos,intervalos,dtype=float)
#b=intervalos_clase(arg2[0],arg2[49],intervalos,9)
print(b,intervalos)
fig, ax = plt.subplots()
intervalos_clase=ax.hist(arg2, bins=b,range=(arg2[0],arg2[99]), linewidth=0.5, edgecolor="white")
plt.xticks(fontsize=8)
ax.set(xlim=(14, 19), xticks=intervalos_clase[1],
       ylim=(0, 33), yticks=np.arange(0, 33, 3,dtype=np.int16))
print(intervalos_clase)
ax.grid(True)
```

```python
listaDatos = arg1.flatten().tolist()

mediaAritmetica = np.mean(arg1)
#Usando numpy
#mediaGeometrica = np.exp(np.mean(np.log(arg1)))
#Usando scipy
#mediaGeometrica = scst.gmean(arg1.flatten())
#usando statistics
mediaGeometrica = sts.geometric_mean(listaDatos)
mediaArmonica = sts.harmonic_mean(listaDatos)

print("La media aritmetica es: " + str(mediaAritmetica))
print("La media geometrica  es: " + str(mediaGeometrica))
print("La media armonica  es: " + str(mediaArmonica))

```

### Conclusiones

De los datos obtenidos podemos concluir que la concentración de personas que ven el anime "La boticaria" tienen una edad media de 16 años, ya que la mayor parte se encuentra en este rango, y por lo tanto esto nos ayuda a poder distribuir de manera correcta y equilibrada el total de la muestra. La media geometrica (la cual nos ayuda a mantener estable nuestra medición debido a que mantiene un enfoque en los datos intermedios) se encuentra en 16 años. La media armónica (la cual se enfoca en los valores más pequeños o minimos de nuestros datos) también nos da un resultado de 16 años. En pocas palabras, considerando todos los tipos de datos, tanto el punto de equilibrio ideal (media aritmética), los datos intermedios (media geométrica) y los datos más pequeños (media armonica), llegamos a la conclusión de que las personas que ven el anime "La botecaria" tienen una edad ideal de 16 años.
